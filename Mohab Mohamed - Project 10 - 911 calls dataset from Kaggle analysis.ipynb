## Dataset source: https://www.kaggle.com/mchirico/montcoalert

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize': [9, 9]}, font_scale=1.2)
```

name of the data file: '911.csv'

```python
# read the csv file and check it out

data = pd.read_csv('911.csv')
```

```python
data.head()
```

```python
# create a plot to check the reason of the call
sns.countplot(x = 'Reason', data = data)
```

what do you make of the plot?

    -   The reason of most of the calls is because of EMS

```python
# create a plot to check the hours in which the calls came
sns.histplot(data['hour'])
```

what do you make of the plot?

    - Most calls came from 6:AM to 10:pm

```python
# create a plot to check the months in which the calls came
sns.countplot(data['month'])
```

what do you make of the plot?

    -  Call are very few on October 

```python
# create a plot to check the day-of-week in whick the calls came
sns.countplot(data['day of week'])
```

what do you make of the plot?

    -   Number of calls is almost the same each day and the least is on Sunday

```python
# create a plot to check the hour of the call and give them a hue depending on the reason

sns.countplot(x = 'hour', hue= 'Reason', data= data)
```

what do you make of the plot?

    -   Roughly from midnight till 6:AM the reason for most calls is EMS, and after 6:AM calls of traffic starts to increase along with EMS calls till 8:PM, then starts to decrease again

```python
# create a plot to check the month of the call and give them a hue depending on the reason
sns.countplot(x = 'month', hue= 'Reason', data= data)
```

what do you make of the plot?

    -  The month of October tend to be the least month of calls for all reasons 

```python
# create a plot to check the day-of-week of the call and give them a hue depending on the reason
sns.countplot(x = 'day of week', hue= 'Reason', data= data)
```

what do you make of the plot?

    -   Not much change for the number of calls over days of the week, but we can say calls are at least on Sunday

```python
# create a plot to check the year of the call and give them a hue depending on the reason
sns.countplot(x = 'Year', hue= 'Reason', data= data)
```

what do you make of the plot?

    -   Calls have increased clearly for all reasons from 2015 to 2016
